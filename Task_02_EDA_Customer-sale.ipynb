### data type conversion
- invice_no and customer_id cannot convert because I am can't perform mathematical operations on these columns
- gender column have limited  unique values so I convert into category for less memory utilization
- Branch column have limited  unique values so I convert into category for less memory utilization
- Customer type column have limited  unique values so I convert into category for less memory utilization
- age column is already in correct numeric data type
- products line are limited so I convert into category for less memory utilization
- quantity already in correct numeric type for mathematical operation 
- payment_method are limited so I am converted into category
 

```python
df['Gender']=df['Gender'].astype('category')
df['Branch']=df['Branch'].astype('category')
df['Customer type']=df['Customer type'].astype('category')
df['City']=df['City'].astype('category')
df['Product line']=df['Product line'].astype('category')
df['Payment']=df['Payment'].astype('category')
```

```python
df.info()
```

### Univariate Analysis

```python
numeric_cols = ['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs',
                'gross margin percentage', 'gross income', 'Rating']

for col in numeric_cols:
    plt.figure(figsize=(12,5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
# Categorical columns
categorical_cols = ['Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Payment']

# Plot for each categorical column
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col], palette='viridis')
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

### Bivariate Analysis

```python
# Scatter plot: Unit price vs Quantity
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Unit price', y='Quantity')
plt.title('Unit Price vs Quantity')
plt.show()
```

```python
# Boxplot: Gender vs Total
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Gender', y='Total')
plt.title('Total by Gender')
plt.show()
```

```python
# Clustered bar plot
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Product line', hue='Gender')
plt.title("Clustered Bar Plot: Gender vs Product Category")
plt.show()
```

### Multivariate analysis 

```python
sns.pairplot(df[['Unit price', 'Quantity', 'Total']])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x='Branch', y='Total', hue='Gender')
plt.title('Total Sales by Branch and Gender')
plt.show()
```

```python
cross_tab = pd.crosstab(df['Gender'], df['Product line'])

# Plot heatmap
plt.figure(figsize=(5, 3))
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu')
plt.title("Heatmap: Gender vs Product Category")
plt.show()
```

### Data Visulization

```python
# 1. Sales by Product Line
plt.figure(figsize=(10,6))
sns.barplot(x='Product line', y='Total', data=df, estimator=sum, ci=None, palette='viridis')
plt.title('Total Sales by Product Line', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Gender-wise Spending
plt.figure(figsize=(7,5))
sns.barplot(x='Gender', y='Total', data=df, estimator=sum, ci=None, palette='coolwarm')
plt.title('Total Sales by Gender', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 3. Payment Method Popularity
plt.figure(figsize=(7,5))
sns.countplot(x='Payment', data=df, palette='Set2')
plt.title('Count of Payment Methods', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 4. Rating distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=10, kde=True, color='skyblue')
plt.title('Customer Rating Distribution', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 5. Branch vs Gross Income
plt.figure(figsize=(7,5))
sns.barplot(x='Branch', y='gross income', data=df, estimator=sum, ci=None, palette='plasma')
plt.title('Gross Income by Branch', fontsize=16)
plt.tight_layout()
plt.show()
```

## 3. Univariate Analysis

- *Plot Used*: countplot

###  Product Line Distribution
- *Plot Used*: countplot
- *Insight: Most popular product line is **Fashion accessories, while **Health and beauty* is least.

###  Payment Methods
- *Plot Used*: countplot
- *Insight: **Ewallet* is the most used payment method, followed by *Cash*.

---

## 4. Bivariate Analysis

###  Gender vs Total Purchase
- *Plot Used*: boxplot
- *Insight: Median spending of **females* is slightly higher than males.

###  Product Line vs Total
- *Plot Used*: boxplot / barplot

###  Payment Method vs Rating
- *Plot Used*: boxplot
- *Insight: Customers using **Credit cards* gave higher ratings on average.

---

## 5. Time-based Analysis

###  Sales by Time of Day
- *Plot Used*: lineplot / histogram
- *Insight: Peak sales occur in the **afternoon*.

###  Day-wise Sales
- *Plot Used*: countplot

## 6. Correlation Analysis

###  Heatmap of Numeric Columns

## 7. Revenue Analysis

### ➤ Branch-wise Revenue
